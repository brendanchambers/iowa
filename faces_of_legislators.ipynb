```python
import json
import urllib.request
from bs4 import BeautifulSoup
```

```python
target_path = 'data/SF359 house vote record.json'
with open(target_path, 'r') as f:  
    house_votes_record = json.load(f)
    
target_path = 'data/SF359 senate vote record.json'
with open(target_path, 'r') as f:
    senate_votes_record = json.load(f)
```

```python
yes_idx = 0
no_idx = 0 # counters for naming the files
didnotvote_idx = 0
nodata_idx = 0

def get_profile_images(records, yes_idx, no_idx, didnotvote_idx, nodata_idx):
    for record in records:
        target_url = 'http://' + record['votesmart_page']
        with urllib.request.urlopen(target_url) as response:
            page = response.read()
            soup = BeautifulSoup(page, 'html.parser')
            profile_img_section = soup.body.find_all('img',{'class':'canphoto lazy'})
            img_postfix = profile_img_section[0]['data-original']
            img_postfix = img_postfix[2:]
            img_url = 'https://' + img_postfix
            print(img_url)
            print(record)

            if record['vote']=='Yea':
                target_file = 'yes_' + str(yes_idx) + '.jpg'
                yes_idx += 1
            elif record['vote']=='Nay':
                target_file = 'no_' + str(no_idx) + '.jpg'
                no_idx += 1
            elif record['vote']=='Did Not Vote':
                target_file = 'didnotvote_' + str(didnotvote_idx) + '.jpg'
                didnotvote_idx += 1
            else:
                print('warning .... no data:   ' + record['vote'])
                target_file = 'nodata_' + str(nodata_idx) + '.jpg'
                nodata_idx += 1

            save_path = 'data/portraits/'  + target_file
            urllib.request.urlretrieve(img_url, save_path)
            
get_profile_images(house_votes_record, yes_idx, no_idx, didnotvote_idx, nodata_idx) 
print('completed gathering house images')

get_profile_images(senate_votes_record, yes_idx, no_idx, didnotvote_idx, nodata_idx) 
print('completed gathering senate images')
        

    
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image

path = 'data/portraits/'

# did not vote
plt.figure(figsize=[16,16])
HEIGHT = 1
WIDTH = 10
idx = 0

for filename in os.listdir(path):
    if filename[0:10] == 'didnotvote':
        img = Image.open(path + filename)

        ax = plt.subplot(HEIGHT, WIDTH ,idx+1)
        ax.imshow(img)
        ax.axis('off')

        idx += 1
    
plt.axis('off')
plt.title('did not vote')
plt.show()


# voted no
plt.figure(figsize=[16,16])
HEIGHT = 6
WIDTH = 10
idx = 0

for filename in os.listdir(path):
    if filename[0:2] == 'no':
        img = Image.open(path + filename)

        ax = plt.subplot(HEIGHT, WIDTH ,idx+1)
        ax.imshow(img)
        ax.axis('off')

        idx += 1
    
plt.axis('off')
plt.title('voted no')
plt.show()


# voted yes
plt.figure(figsize=[16,16])
HEIGHT = 6
WIDTH = 10
idx = 0

for filename in os.listdir(path):
    if filename[0:3] == 'yes':
        img = Image.open(path + filename)

        ax = plt.subplot(HEIGHT, WIDTH ,idx+1)
        ax.imshow(img)
        ax.axis('off')

        idx += 1
    
plt.axis('off')
plt.title('voted yes')
plt.show()
```
